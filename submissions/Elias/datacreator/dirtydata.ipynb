```python
df.loc[df.sample(frac=0.1).index, 'gross_income'] = np.nan
```

```python
df = pd.concat([df, df.sample(n=10)], ignore_index=True)
```

```python
df.loc[df.sample(n=5).index, 'unit_price'] = 5000
```

```python
indices = df.sample(frac=0.1).index
# 2. Convertir números a texto agregando un símbolo (ej: 45.0 -> "45.0 $")
df.loc[indices, 'Unit price'] = df.loc[indices, 'Unit price'].astype(str) + ' $'
```

```python
df.loc[df.sample(frac=0.05).index, 'Customer type'] = 'Memmber'
df.loc[df.sample(frac=0.05).index, 'City'] = 'Yangoon'
```

```python
# Introduce 'Bitcoin' en el 5% de los pagos (el supermercado no acepta criptos)
df.loc[df.sample(frac=0.05).index, 'Payment'] = 'Bitcoin'

# Introduce '?' en el 5% de la columna de género
df.loc[df.sample(frac=0.05).index, 'Gender'] = '?'
```

```python
# Convertimos toda la columna de Cantidad e Ingresos a texto (string)
df['Quantity'] = df['Quantity'].astype(str)
df['gross_income'] = df['gross_income'].astype(str)
```

```python
# Diccionario para mapear los nombres correctos a nombres 'sucios'
nombres_sucios = {
    'InvoiceID': 'Invoice ID ',    # Espacio invisible al final (causa KeyErrors)
    'City': 'CITY',                # Inconsistencia (Gritando en mayúsculas)
    'Unit price': 'unit_price',    # Cambio de estilo (Snake case vs Camel case)
    'Tax 5%': 'Tax(5%)'            # Paréntesis que complican la sintaxis
}

# Aplicamos los cambios al dataset
df.rename(columns=nombres_sucios, inplace=True)
```

```python
# Selecciona el 10% de las filas al azar
indices_sucios = df.sample(frac=0.1).index

# Convierte el número a texto y le pega el símbolo (ej: 26.1 -> "26.1€")
df.loc[indices_sucios, 'gross_income'] = df.loc[indices_sucios, 'gross_income'].astype(str) + '€'
```
