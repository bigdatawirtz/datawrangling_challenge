```python
import pandas as pd
import numpy as np
import random
import io

df = pd.read_csv('world-happiness-report-2019.csv', encoding='utf-8')
print("--- Dataframe Original Cargado ---")
print(df.info())

np.random.seed(42)
random.seed(42)

mask = np.random.random(df.shape) < 0.15
columnas_numericas = df.select_dtypes(include=[np.number]).columns
df[columnas_numericas] = df[columnas_numericas].mask(mask[::, 1:], other=np.nan)

duplicates = df.sample(n=5)
df = pd.concat([df, duplicates], ignore_index=True)
df.loc[len(df)] = df.iloc[0].copy()
df.iloc[-1, 1] = 9999


indices_outliers = np.random.choice(df.index, 3, replace=False)
df.loc[indices_outliers, 'Ladder'] = df.loc[indices_outliers, 'Ladder'] * 100
df.loc[np.random.choice(df.index), 'Social support'] = -500

indices_format = np.random.choice(df.index, 5, replace=False)
df['Positive affect'] = df['Positive affect'].astype(object)
for idx in indices_format:
    val = df.loc[idx, 'Positive affect']
    if pd.notnull(val):
        df.loc[idx, 'Positive affect'] = str(val).replace('.', ',') + " pts"

def ensuciar_texto(texto):
    if not isinstance(texto, str) or len(texto) < 4:
        return texto
    lista = list(texto)
    i = random.randint(0, len(lista) - 2)
    lista[i], lista[i+1] = lista[i+1], lista[i]
    return "".join(lista)

indices_typo = np.random.choice(df.index, 10, replace=False)
df.loc[indices_typo, 'Country (region)'] = df.loc[indices_typo, 'Country (region)'].apply(ensuciar_texto)

df.loc[len(df)] = [np.nan] * len(df.columns)
df.iloc[-1, 0] = "Atlantis_Region_X"
df.iloc[-1, 1] = 500

df['Freedom'] = df['Freedom'].astype(str)
idx_str = np.random.choice(df.index)
df['Corruption'] = df['Corruption'].astype(object)
df.loc[idx_str, 'Corruption'] = "High"

print("\n--- Dataframe Sucio Generado ---")
print(df.info())
print("\n--- Muestra de los datos sucios (últimas 20 filas) ---")
print(df.tail(20))

# Guardar a CSV
df.to_csv("dirty_happiness_data.csv", index=False)
print("\nArchivo guardado como 'dirty_happiness_data.csv'")
```
