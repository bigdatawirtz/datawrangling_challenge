Hacer la conversión despues de colocar los tipos en nulo sería mucho más destructivo y problematico, isna() no lo detecta porque en NaN se convierte al literal "nan". No lo vamos ha hacer porque sería muy molesto para los compañeros.

```python
duplicates = pd.DataFrame([df.iloc[rd.randint(0, df.shape[0]-1)].values for x in range (0, rd.randint(1,5))], columns=df.columns)
df = pd.concat([df, duplicates])
```

```python
for column in df.columns:
    if (rd.randint(0,10)<7):
        for k in df.loc[:, column].keys(): 
            if (rd.randint(0,10)<5): df.loc[k, column] = None
    if (rd.randint(0,10)<3):
        for k in df.loc[:, column].keys():
            if (rd.randint(0,100)<10):
                if isinstance(k, (int, float)): df.loc[k, column] += 1000
                if isinstance(k, (str)): k += "%"
```
